```python
import pytesseract
from PIL import Image
import os


# Set the path to the folder containing all the images
images_folder_path = "/content/dataset/train/images"

# Function to perform OCR on an image and return the extracted text
def perform_ocr(image_path):
    img = Image.open(image_path)
    extracted_text = pytesseract.image_to_string(img)
    return extracted_text

# List all image files in the folder
image_files = [f for f in os.listdir(images_folder_path) if f.endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]

# Create an empty list to store the extracted text from all images
all_extracted_text = []

# Loop through each image file and perform OCR
for image_file in image_files:
    image_path = os.path.join(images_folder_path, image_file)
    extracted_text = perform_ocr(image_path)
    all_extracted_text.append(extracted_text)

# Print the extracted text from all images
for i, extracted_text in enumerate(all_extracted_text, 1):
    print(f"Extracted Text from Image {i}:")
    print(extracted_text)
    print("--------")
```

```python
from spacy.matcher import Matcher
import spacy

# Load the pre-trained English model from spaCy
nlp = spacy.load("en_core_web_sm")

# Define a more flexible pattern to extract employer names
matcher = Matcher(nlp.vocab)
employer_name_pattern = [{"POS": "PROPN", "ENT_TYPE": {"NOT_IN": ["ORG"]}}]
matcher.add("EMPLOYER_NAME", [employer_name_pattern])

# Actual extracted text list from OCR (replace this with your actual data)
extracted_text_list = [
    "Employer Name: XYZ Company\nAddress: 123 Main St, Anytown, USA\nSSN: 123-45-6789\nWages: $5000",
    "Employer Name: ABC Corp\nAddress: 456 Oak Ave, Cityville, USA\nSSN: 987-65-4321\nWages: $4500",
    # Add more extracted text entries for other images
]

# Function to extract specified entities from the text
def extract_entities(text):
    doc = nlp(text)

    # Initialize dictionary to store the extracted entities
    entities = {
        "employerName": [],
    }

    # Use the Matcher to find employer names in the text
    matches = matcher(doc)
    for match_id, start, end in matches:
        entities["employerName"].append(doc[start:end].text)

    return entities

# Loop through each extracted text and extract entities
for i, text in enumerate(all_extracted_text, 1):
    entities = extract_entities(text)

    # Print the extracted entities for each text entry
    print(f"Entities from Extracted Text {i}:")
    print(entities)
    print("--------")

```
